## Lists: Accessing Elements
- Individual list elements can be retrieved by providing the position of the item to be retrieved in square brackets [ ]
>        0     1     2
> list1 = ['first','second','third']
>       -3    -2    -1
- Indexing starts at 0
- Negative indexes go backwards from the end
- Individual list elements can be retrieved by providing the position of the item to be retrieved in square brackets [ ]

```python
list1 = ['first','second','third']
print(list1[0])
print(list1[-1])
```

```python
list3 = ['pi', 3.1415925, 'answer', 42] 
print("%s is roughtly %f" % (list3[0],list3[1]))
```

## Lists: Appending and Extending
- An item can be appended to the end of a list using append

```python
print(len(list1))
list1.append('fourth')
print(list1[-1])
print(len(list1))
```

- A list of items can be appended to the end of a list using extend

```python
list1.extend(["fifth","sixth"])
print(list1)
```

## Lists: Slicing
- A copy of a sublist can be created using the slice operator
> list[start:end]
- The elements at the index start is included, the one at end is excluded <br>
- Both indices can be ommitted, defaulting to the start and end of the list respectively <br>

```python
print(list1[0:1])
print(list1[1:3])
print(list1[:2])
print(list1[1:])
print(list1[:])
```

### Advanced (optional)

Plot relative humidity data from NCEP reanalysis data.
xarray.open_dataset can open and decode a dataset from a file path or an OpenDAP URL.

The dataset that we will be using is here:

```python
import xarray as xr
file1 = 'http://www.esrl.noaa.gov/psd/thredds/dodsC/Datasets/ncep.reanalysis/surface/rhum.sig995.2021.nc'

# this automatically closes the dataset after use
with xr.open_dataset(file1) as ds1:
    print(ds1)
    rh=ds1.rhum
```

Plot relative humidity for the last time step. This time let's define air temperature as the following:
air = ds.air

```python
rh.isel(time=-1).plot()
```

Now let's calculate specific humidity.

Using relative humidity (rh), specific humidity (q) can be calculated as:

q = qs * rh /100 

where qs is specific humidity at saturation. We divide the value by 100 because rh is in percent in the data. Let's first calculate qs.

qs = (0.622 * es)/(p - (0.378 * es)) </br>

where p is pressure, and es is saturation vapor pressure.


Using the Clausius-Clapeyron relation, es can be calculated as the following: </br>

es= 0.611 * exp ( 17.502 * T / ( (T  + 240.97 ) ) ) </br>

where T is temperature in Celcius. This is probably the only equation that uses temperature in Celcius. <br>

For the programming, let's follow these steps in reverse order.

```python
# We need numpy to calculate exp values.
import numpy as np
```

Let's calculate es.
We need air temperature.
Let's read air temperature from the NCEP data.

```python
file2 = 'http://www.esrl.noaa.gov/psd/thredds/dodsC/Datasets/ncep.reanalysis/surface/air.sig995.2021.nc'

# Let's define T
with xr.open_dataset(file2) as ds2:
    print(ds2)
    t=ds2.air-273.15 # T in C
```

```python
# Let's take a look at T for the last time step.
t.isel(time=-1).plot()
```

```python
# Let's calculate es and qs
es = 0.611 * np.exp ( 17.502 * t / ( (t + 240.97 ) ) )*1000. # We multiply 1000 to convert kPa to Pa.
```

```python
# To convert es to qs, we need pressure. All the values are at surface.
file3 = 'http://www.esrl.noaa.gov/psd/thredds/dodsC/Datasets/ncep.reanalysis/surface/pres.sfc.2021.nc'
with xr.open_dataset(file3) as ds3:
    print(ds3)
    ps=ds3.pres
```

```python
# Let's take a look at ps for the last time step.
ps.isel(time=-1).plot()
```

```python
# Let's calculate qs
qs = (0.622 * es)/(ps - (0.378 * es)) 
print(qs.shape)
print(ps.shape)
print(es.shape)
```

```python
# Calculate q
q=qs*rh/100.*1000.   # division by 100: from percent to value; multiplication of 1000.: from kg/kg to g/kg
```

```python
# Let's take a look at ps for the last time step.
q.isel(time=-1).plot()
```
