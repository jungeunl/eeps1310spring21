```python
filename='/Users/jungeunl/Cycad/class/data/cru_ts4.03.1901.2018.tmp.dat.nc' #using Python on my laptop
```

```python
with xr.open_dataset(filename) as ds:
    print(ds)
```

```python
lev1=np.arange(-40,40,1)
ds.tmp.isel(time=-1).plot(levels=lev1)
```

```python
#Plot temperature of a point in January
lat_pvd=41.8
lon_pvd=-71.4
t_pvd=ds.tmp.sel(lat=lat_pvd,lon=lon_pvd, method='nearest')
t_pvd.plot()
```

```python
t_roll_pvd = t_pvd.rolling(time=120, center=True).mean() # 10 year running mean
t_roll_pvd.plot()
```
